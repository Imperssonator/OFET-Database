```python
import numpy as np
from scipy.io import loadmat  # this is the SciPy module that loads mat-files
import matplotlib.pyplot as plt
from datetime import datetime, date, time
import pandas as pd
import bokeh, bokeh.plotting
import IPython
import ipywidgets
import qgrid

mat = loadmat('/Users/Imperssonator/Documents/Matlab/OFET-Database/Functions/OFETDatabase.mat')  # load mat-file
```

```python
rows = []
for row in mat['OFET'][0]:
    tmp = []
    for v in [v[0].tolist() for v in row]:
        if isinstance( v, list):
            tmp.append(v[0])
        else:
            tmp.append(v)
    rows.append(tmp)
```

```python
df = pd.DataFrame(rows,columns=mat['OFET'].dtype.names)
```

```python
qgrid.show_grid(df, grid_options={'forceFitColumns': False, 'defaultColumnWidth': 100})
```
